```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
latihan2 = pd.read_excel("../../Data/latihan2.xlsx")
latihan2
```

```python
# Ubah data kategorikal menjadi numerikal

replace_map = {
    'Hasil Sebenarnya': {'Lulus': 0, 'Tidak Lulus': 1},
    'Hasil Prediksi': {'Lulus': 0, 'Tidak Lulus': 1}
}

latihan2.replace(replace_map, inplace=True)
latihan2
```

```python
# split actual and predicted values
actual = latihan2["Hasil Sebenarnya"].astype(int)
predicted = latihan2["Hasil Prediksi"].astype(int)

# Confusion Matrix
cm = confusion_matrix(actual, predicted)

df_cm = pd.DataFrame(cm,
                    index=["Actual Lulus", "Actual Tidak Lulus"],
                    columns=["Pred Lulus", "Pred Tidak Lulus"])

print("\nConfusion Matrix:")
print(df_cm)
```

```python
# Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Lulus', 'Tidak Lulus'], yticklabels=['Lulus', 'Tidak Lulus'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Accuracy, Precision, dan Recall
accuracy = accuracy_score(actual, predicted)
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
